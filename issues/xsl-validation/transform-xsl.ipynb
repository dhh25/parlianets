```python
# based on https://pypi.org/project/saxonche/

from saxonche import *

CORPUSPATH = "/home/simon/parliamint/ParlaMint-BG-en.TEI.ana/"
STYLESHEETPATH = "/home/simon/parlianets/issues/"


proc = PySaxonProcessor(license=False)
 
xsltproc = proc.new_xslt30_processor()
document = proc.parse_xml(xml_file_name=CORPUSPATH + "2014/ParlaMint-BG-en_2014-10-27.ana.xml")
#document = proc.parse_xml(xml_text="<doc><item>text1</item><item>text2</item><item>text3</item></doc>")
executable = xsltproc.compile_stylesheet(stylesheet_file=STYLESHEETPATH + "parlamint2meta.xsl")


value = PyXdmValue(proc)
value.add_xdm_item(proc.make_string_value(CORPUSPATH + "ParlaMint-BG-en.ana.xml"))
executable.set_parameter("meta", value)

output = executable.transform_to_string(xdm_node=document)
print(output)
```
